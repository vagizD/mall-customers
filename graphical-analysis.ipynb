# Mall customers analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data import and first view:
mall_customers = pd.read_csv('Mall_Customers.csv', index_col='CustomerID')
mall_customers.head()
```

### Creating plot to visualize age-score and income-score trends for males and females.

```python
fig, (ax1, ax2) = plt.subplots(nrows=1,
                               ncols=2,
                               figsize=(18, 6))

fig.suptitle('Age-score and income-score trends for males and females', fontsize=12, fontweight='bold')

x = mall_customers['Age']
y = mall_customers['Spending Score (1-100)']
c = mall_customers['Genre']


males = np.array([])
females = np.array([])
for i in c.index:
    if c[i] == "Male":
        males = np.append(males, i)
    else:
        females = np.append(females, i)

ax1.scatter(x[males], y[males], color='cyan', edgecolors='black', label='Male')
ax1.scatter(x[females], y[females], color='red', edgecolors='black', label='Female')
ax1.axhline(y[males].mean(), color='blue', linestyle='--', label='Male Avg')
ax1.axvline(x[males].mean(), color='blue', linestyle='--')
ax1.axhline(y[females].mean(), color='purple', linestyle='--', label='Female Avg')
ax1.axvline(x[females].mean(), color='purple', linestyle='--')

ax1.set(title='Age & Score',
        xlabel='Age',
        ylabel='Score')

ax1.grid(b=True, linestyle='-.', linewidth=0.7, alpha=0.8)
ax1.set_xlim([10, 80])
ax1.set_ylim([0, 100])

legend2 = ax1.legend(title='GENRE', loc='upper right')

x = mall_customers['Annual Income (k$)']*1000
y = y

ax2.scatter(x[males], y[males], color='cyan', edgecolors='black', label='Male')
ax2.scatter(x[females], y[females], color='red', edgecolors='black', label='Female')
ax2.axhline(y[males].mean(), color='blue', linestyle='--', label='Male Avg')
ax2.axvline(x[males].mean(), color='blue', linestyle='--')
ax2.axhline(y[females].mean(), color='purple', linestyle='--', label='Female Avg')
ax2.axvline(x[females].mean(), color='purple', linestyle='--')

ax2.set(title='Annual Income & Score',
        xlabel='Annual Income, $',
        ylabel='Score')

ax2.grid(b=True, linestyle='-.', linewidth=0.7, alpha=0.8)
ax2.set_xlim([0, 140000])
ax2.set_ylim([0, 100])

ax2.legend(title='GENRE', loc='upper right');
```

### Analysis.

On both graphs, females and males are very close to each other (on average). An average mall customer score is 50, annual pay is $60 000, and age is 39.

```python
mall_customers.describe()
```

Age-Score plot has high variance and data is spreaded. Income-Score plot has strongly marked point of preferrence. Score of 50 is common among people with Annual Income (AI) of \\$42 000 - \\$64 000, whereas very uncommon among people beyond that AI range.

There could be made generalisations about data based on two graphs. Talking about first graph, noone aged more than 40 has a score higher than 60.

```python
over_40 = mall_customers[mall_customers['Age'] > 40]
over_40.describe()
```

```python
def addlabels(axes, x, y, labels):
    for i in range(len(x)):
        axes.text(i, y[i], labels[i], ha='center')

fig, ax = plt.subplots(figsize=(7,4))

y = np.array([])
for genre in [males, females]:
    mean = len(genre)/len(mall_customers.index)
    y = np.append(y, mean)
x = ['Male', 'Female']

ax.bar(x=x, height=y, color='cyan', width=0.5)
plt.yticks([])

addlabels(ax, x, y, [f"{int(mean*100)}%" for mean in y])
ax.set(title='Genre Distribution');
```
